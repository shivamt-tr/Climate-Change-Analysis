```python
import numpy as np
import pandas as pd
import os
```

```python
first_file = True
for file in os.listdir('./../datasets/raw-datasets/GHG_Dataset/'):

    file_name = file[:-4].split("-")
    gas = file_name[-1]
    sector = " ".join(file_name[2:-1])
    df = pd.read_csv("./../datasets/raw-datasets/GHG_Dataset/" + str(file))
    df.dropna(axis = 0, inplace = True)
    df.replace("false", 0, inplace = True)
    df["Sector"] = sector
    df["Gas"] = gas
    df.drop("unit", axis = 1, inplace=True)
    
    if first_file:
        first_file = False
        main_df = df
    else:
        main_df = main_df.append(df)
```

```python
main_df
```

```python
col = list(map(str, list(range(1990, 2019))))
main_df[col] = main_df[col].astype("float")
```

```python
gb_gas = main_df.groupby(by = "Gas").sum()
gb_gas.sort_values(by = col[-1::-1], ascending=False, inplace=True)
gb_gas.T
```

```python
gb_country = main_df.groupby(by = "Country/Region").sum()

world = np.round(np.array(gb_country.iloc[:, :].sum(axis=0)), 2)
world = np.reshape(world, (29,1))
df = pd.DataFrame(world).T
column = list(map(str, list(range(1990, 2019))))
df.columns = column
gb_country = gb_country.append(df)

as_list = gb_country.index.tolist()
idx = as_list.index(0)
as_list[idx] = 'World'
gb_country.index = as_list
gb_country.index.name = "Country"

gb_country.sort_values(by = col[-1::-1], ascending=False, inplace=True)

gb_country
```

```python
gb_sector = main_df.groupby(by = "Sector").sum()
gb_sector.sort_values(by = col[-1::-1], ascending=False, inplace=True)
gb_sector.T
```

```python
main_df.to_csv("./../datasets/cleaned-datasets/GHG_Complete.csv", index = False)
gb_gas.to_csv("./../datasets/cleaned-datasets/GHG_Gas.csv")
gb_country.to_csv("./../datasets/cleaned-datasets/GHG_Country.csv")
gb_sector.to_csv("./../datasets/cleaned-datasets/GHG_Sector.csv")
```
