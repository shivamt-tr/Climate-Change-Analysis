```python
change = [("Bahamas, The", "Bahamas"), ("Brunei Darussalam" , "Brunei"), ("Cabo Verde", "Cape Verde"), 
          ("Cote d'Ivoire", "CÃ´te d'Ivoire"),("Congo, Dem. Rep.", "Democratic Republic of the Congo"), 
          ("Egypt, Arab Rep.", "Egypt"), ("Gambia, The", "Gambia"), ("Iran, Islamic Rep.", "Iran"), 
          ("Kyrgyz Republic", "Kyrgyzstan"), ("Lao PDR", "Laos"), ("Micronesia, Fed. Sts.", "Micronesia"),
          ("North Macedonia", "Macedonia"), ("Korea, Dem. People's Rep.", "North Korea"), ("Congo, Rep.", "Republic of Congo"), 
         ("Russian Federation", "Russia"), ("St. Kitts and Nevis", "Saint Kitts and Nevis"), ("St. Lucia", "Saint Lucia"), 
         ("St. Vincent and the Grenadines", "Saint Vincent and the Grenadines"), ("Slovak Republic", "Slovakia"), 
         ("Korea, Rep.", "South Korea"), ("Syrian Arab Republic", "Syria"), ("Venezuela, RB", "Venezuela"), 
         ("Yemen, Rep.", "Yemen")]
```

```python
for ch in change:
    df_pop.replace(ch[0], ch[1], inplace=True)
```

```python
ghg_country = set(df_ghg["Country"].unique())
pop_country = set(df_pop["Country"].unique())
unique_country = pop_country.intersection(ghg_country)
len(unique_country)
```

```python
drop_country = pop_country.difference(ghg_country)
```

```python
for country in drop_country:
    df_pop = df_pop[df_pop["Country"] != country]
```

```python
country_code = dict()
```

```python
for country in unique_country:
    code = df_pop[df_pop["Country"] == country]["Country Code"].iloc[0]
    country_code[country] = code
```

```python
with open("./../datasets/cleaned-datasets/country_code.json", "w") as outfile:
    json.dump(country_code, outfile)
```

```python
drop_column = list(map(str, list(range(1960, 1990))))
drop_column.append("2019")
drop_column.append("2020")
drop_column.append("Country Code")
```

```python
df_pop.drop(drop_column, axis = 1, inplace=True)
```
